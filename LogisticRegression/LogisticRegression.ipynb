```python
import numpy as np
```

```python
# Sigmoid Function
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
```

```python
# Calculate the gradient
def calculate_gradient(theta, X, y):
  m = y.size # number of features
  result = (X.T @ (sigmoid(theta @ X) - y)) / m
  return result
```

```python
# Gradient Descent
def gradient_descent(theta, X, y, alpha = 0.1, num_iterations=100, tolerance = 1e-7):
  x_b = np._c[np.ones((X.shape[0], 1)), X] # Add the bias vector to the calculation.
  theta = np.zeros[x_b.shape[1]]

  # Continously update the parameters (theta)
  for i in range(num_iterations):
    grad = calculate_gradient(theta, x_b, y)
    theta -= alpha * grad

    # Gradient must stay above the tolerance level
    if np.linalg.norm(grad) < tolerance:
      break

  return theta
```

```python
# Predict the Probability
def predict_proba(X, theta):
  X_b = np._c[np.ones((X.shape[0], 1)), X] # Add the bias vector to the calculation.
  return sigmoid(X_b @ theta)
```

```python
# Make the prediction.
def predict(X, theta, threshold = 0.5):
  return (predict_proba(X, theta) >= threshold).astype(int)
```
